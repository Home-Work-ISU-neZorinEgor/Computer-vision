```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_names = [f"figure{i}.txt" for i in range(1, 7)]

for file_name in file_names:
    with open(file_name, "r") as file:
        lines = file.read().splitlines()

    max_size_mm = float(lines[0])
    image_data = np.array([list(map(int, line.split())) for line in lines[2:]])
    nominal_resolution = max_size_mm / image_data.shape[1]

    print(f"Для файла {file_name} Номинальное разрешение: {nominal_resolution:.2f} мм/пиксель", end="\n")

    plt.imshow(image_data, cmap='gray')
    plt.title(f"Изображение из файла {file_name}")
    plt.colorbar()
    plt.show()
```
